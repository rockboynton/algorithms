```python
import graphs

import random
import time 
import numpy as np
import matplotlib.pyplot as plt

training, testing = graphs.load_data('facebook_data/training_set.tsv', 'facebook_data/testing_set.tsv')
```

```python
# Evaluate how the number of recommendations / run time varies with the max depth.
max_depths = []
num_edges = []
run_times = []
for max_depth in range(1, 11):
    max_depths.append(max_depth)
    start = time.clock()
    H = graphs.recommend_all_friends(training, max_depth)
    end = time.clock()
    num_edges.append(H.count_edges())
    run_times.append(end - start)

plt.plot(max_depths, num_edges)
plt.xlabel('Max Depth', fontsize=18)
plt.ylabel('Number of Edges', fontsize=18)
plt.title('Number of Recommendations vs Max Depth')
plt.show()

plt.plot(max_depths, run_times)
plt.xlabel('Max Depth', fontsize=18)
plt.ylabel('Run Time', fontsize=18)
plt.title('Run time vs Max Depth')
plt.show()
```
