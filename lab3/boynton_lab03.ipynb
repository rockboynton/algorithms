## Introduction

In this notebook, the Heap Sort algorithm is going to benchmarked. in particular, the insertion and deletion times of Heap Sort willl be compared for sorted, reverse sorted, and randomly-shuffled lists of numbers.

 ## Summary of Results



---

```python
import sorting

import random
import time 
import numpy as np
import matplotlib.pyplot as plt


def benchmark_build(case):
    total_time_avgs = []
    numbers = [100, 1000, 10000, 100000]
    n_trials = 10
    for n in numbers:
        times = []
        for i in range(n_trials):
            lst = [random.random() for _ in range(n)]
            if case == 'random':
                random.shuffle(lst)
            elif case == 'sorted':
                list.sort(lst)
            elif case == 'reverse sorted':
                list.sort(lst)
                list.reverse(lst)
            start = time.clock()
            sorting.build_max_heap(lst)
            end = time.clock()
            elapsed = end - start
            times.append(elapsed)
        total_time_avgs.append(np.mean(times))
    plt.plot(np.log10(numbers), total_time_avgs, label=f'Building Max Heap ({case.capitalize()})')
    plt.xlabel('List Size (log10)', fontsize=18)
    plt.ylabel('Running Time (s)', fontsize=18)
    plt.legend()


benchmark_build(case='random')
benchmark_build(case='sorted')
benchmark_build(case='reverse sorted')
plt.title('Impact of List Order on Performance')
```

```python
def benchmark_build_vs_insert(case):
    build_total_time_avgs = []
    insert_total_time_avgs = []
    numbers = [100, 1000, 10000, 100000]
    n_trials = 10
    for n in numbers:
        build_times = []
        insert_times = []
        for i in range(n_trials):
            build_lst = [random.random() for _ in range(n)]
            if case == 'random':
                random.shuffle(build_lst)
            elif case == 'sorted':
                list.sort(build_lst)
            elif case == 'reverse sorted':
                list.sort(build_lst)
                list.reverse(build_lst)
            insert_lst = build_lst[:]

            start = time.clock()
            sorting.build_max_heap(build_lst)
            end = time.clock()
            elapsed = end - start
            build_times.append(elapsed)

            heap = []
            start = time.clock()
            for num in insert_lst:
                sorting.max_heap_insert(heap, num)
            end = time.clock()
            elapsed = end - start
            insert_times.append(elapsed)

        build_total_time_avgs.append(np.mean(build_times))
        insert_total_time_avgs.append(np.mean(insert_times))
    plt.plot(np.log10(numbers), build_total_time_avgs, label=f'Build Max Heap ({case.capitalize()})')
    plt.plot(np.log10(numbers), insert_total_time_avgs, label=f'Max Heap Insert ({case.capitalize()})')
    plt.xlabel('List Size (log10)', fontsize=18)
    plt.ylabel('Running Time (s)', fontsize=18)
    plt.legend()
    plt.title(f'Impact of Building Heap in One Step vs Inserting Each Element ({case.capitalize()})')
    plt.show()


benchmark_build_vs_insert(case='random')
benchmark_build_vs_insert(case='sorted')
benchmark_build_vs_insert(case='reverse sorted')
```

```python
def benchmark_extract_vs_insert(case):
    extract_total_time_avgs = []
    insert_total_time_avgs = []
    numbers = [100, 1000, 10000, 100000]
    n_trials = 10
    for n in numbers:
        extract_times = []
        insert_times = []
        for i in range(n_trials):
            build_lst = [random.random() for _ in range(n)]
            if case == 'random':
                random.shuffle(build_lst)
            elif case == 'sorted':
                list.sort(build_lst)
            elif case == 'reverse sorted':
                list.sort(build_lst)
                list.reverse(build_lst)
            insert_lst = build_lst[:]

            sorting.build_max_heap(build_lst)
            start = time.clock()
            while build_lst:
                sorting.heap_extract_max(build_lst)
            end = time.clock()
            elapsed = end - start
            extract_times.append(elapsed)
            
            heap = []
            start = time.clock()
            for num in insert_lst:
                sorting.max_heap_insert(heap, num)
            end = time.clock()
            elapsed = end - start
            insert_times.append(elapsed)

        extract_total_time_avgs.append(np.mean(extract_times))
        insert_total_time_avgs.append(np.mean(insert_times))
    plt.plot(np.log10(numbers), extract_total_time_avgs, label=f'Max Heap Extract({case.capitalize()})')
    plt.plot(np.log10(numbers), insert_total_time_avgs, label=f'Max Heap Insert ({case.capitalize()})')
    plt.xlabel('List Size (log10)', fontsize=18)
    plt.ylabel('Running Time (s)', fontsize=18)
    plt.legend()
    plt.title(f'Max Heap Extraction vs Insert ({case.capitalize()})')
    plt.show()


benchmark_extract_vs_insert(case='random')
benchmark_extract_vs_insert(case='sorted')
benchmark_extract_vs_insert(case='reverse sorted')
```
